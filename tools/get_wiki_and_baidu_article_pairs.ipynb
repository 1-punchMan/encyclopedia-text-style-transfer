```python
import json, linecache, re, json
from opencc import OpenCC
cc_t2s = OpenCC('t2s')
cc_s2t = OpenCC('s2t')
```

```python
def get_wiki_arts(file):
    with open(file, 'r', encoding="utf-8") as f:
        wiki_arts = json.load(f)
    return wiki_arts

def get_baidu_art(file, info):
    return ''.join([linecache.getline(file, i) for i in range(info[0] + 1, info[1])])    # linecache is 1-based.

def get_baidu_arts(metadata_file, arts_file, n):
    baidu_arts = {}
    for i in range(n):
        file = metadata_file.format(n=i + 1)
        with open(file, 'r', encoding="utf-8") as f:
            metadata = json.load(f)
            
        file = arts_file.format(n=i + 1)
        arts = {title: get_baidu_art(file, info[1]) for title, info in metadata.items()}
        baidu_arts.update(arts)
    return baidu_arts
```

```python
metadata_file = "/home/zchen/encyclopedia-text-style-transfer/data/ETST/baidu/processed_data/cleaned/metadata{n}.json"
arts_file = "/home/zchen/encyclopedia-text-style-transfer/data/ETST/baidu/processed_data/cleaned/articles{n}"
baidu_arts = get_baidu_arts(metadata_file, arts_file, 4)
```

```python
def strip_parentheses(s):
    matched = re.search(r"[（(]", s)
    if matched:
        ind = matched.start()
        s = s[:ind]
    return s.strip()

def strip_title(s):
    s = s[2:-2]
    ind = s.find('|')
    if ind == -1:
        ind = None
    s = s[:ind]
    return strip_parentheses(s)

matched = re.findall(r"\[\[.*?\]\]", src)
titles = [strip_title(m) for m in matched]
print(len(titles))

titles = set(titles)
print(len(titles))
print(titles)
```

```python
def get_arts_from_dict(d, st_titles):
    return {k: d[k] for k in d for t in st_titles if t == strip_parentheses(k)}

def get_pair(d1, d2, title):
    print(title)
    st_titles = [title, cc_t2s.convert(title), cc_s2t.convert(title)]
    d1_arts = get_arts_from_dict(d1, st_titles)
    d2_arts = get_arts_from_dict(d2, st_titles)
    return (d1_arts, d2_arts) if d1_arts and d2_arts else None
            
art_pairs = {}
for title in titles:
    pair = get_pair(wiki_arts, baidu_arts, title)
    if pair is not None:
        art_pairs[title] = pair
```

```python
print(len(art_pairs))
for i, (title, pair) in enumerate(art_pairs.items()):
    if i == 0:
        print(title)
        for encycl, arts in zip(["wiki", "baidu"], pair):
            print(encycl)
            for t, text in arts.items():
                print(t)
                print(text)
                print()
            print()
        break
```
